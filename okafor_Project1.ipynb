## The Fisher’s Iris dataset consists of 50 samples from each of three species of Iris (setosa, virginica and versicolor). To simplify the project, just use the data for setosa and versicolor species. Four features were measured from each sample: the length and the width of the sepals and petals, in centimeters. Based on the combination of these four features, Fisher developed a linear discriminant model to distinguish the species from each other.

```python
#include all the modules here
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
import math
```

```python
#To import the Iris dataset:
iris_data = pd.read_csv("iris_data.csv")
# the iris dataset is now a Pandas DataFrame

#To view Iris data below:
iris_data
```

```python
# Show samples from each species
iris_data["species"].value_counts()
```

```python
# Create a dataframe for only setosa and versicolor
iris_data_2 = iris_data[(iris_data.species =='setosa') | (iris_data.species == 'versicolor')]
iris_data_2
```

```python
# Show pairwise relationships in this dataset using seaborn's pairplot 
sns.pairplot(data = iris_data_2, hue = 'species')
```

### choose 2 with the highest distance  petal_lenght vs. petal_width and petal_length and sepal_width

```python
# Get all data for each specimens
setosa = iris_data_2[iris_data_2.species == 'setosa']
versicolor = iris_data_2[iris_data_2.species == 'versicolor']

#Gets the corresponding length and width foe each
setosa_petal_len = np.array(setosa["petal_length"])
setosa_sepal_wid = np.array(setosa["sepal_width"])
versi_petal_len = np.array(versicolor["petal_length"])
versi_sepal_wid = np.array(versicolor["sepal_width"])
setosa_petal_wid = np.array(setosa["petal_width"])
versi_petal_wid = np.array(versicolor["petal_width"])

#Gets the mean for each corresponding lenght an width
setosa_petal_len_mean = setosa_petal_len.mean()
setosa_sepal_wid_mean = setosa_sepal_wid.mean()
versi_petal_len_mean = versi_petal_len.mean()
versi_sepal_wid_mean = versi_sepal_wid.mean()
setosa_petal_wid_mean= setosa_petal_wid.mean()
versi_petal_wid_mean = versi_petal_wid.mean()


# graph points for the mean of petal length vs petal width for both specimens
# setosa1 = [setosa_petal_len_mean, setosa_sepal_wid_mean]
# versicolor1 = [versi_petal_len_mean,versi_sepal_wid_mean]
# setosa2  = [setosa_petal_len_mean, setosa_petal_wid_mean]
# versicolor2 = [versi_petal_len_mean, versi_petal_wid_mean]


# gets the distance between the two means on the graph
dist_versi_setosa1 = math.sqrt((versi_petal_len_mean - setosa_petal_len_mean)**2 + (versi_sepal_wid_mean - setosa_sepal_wid_mean)**2)
dist_versi_setosa2 = math.sqrt((versi_petal_len_mean - setosa_petal_len_mean)**2 + (versi_petal_wid_mean - setosa_petal_wid_mean)**2)

print(dist_versi_setosa1, dist_versi_setosa2)
species = iris_data_2["species"]

```

```python
# choose 2 with the highest distance  petal_lenght vs. petal_width and petal_length and sepal_width
# draw a subplot for two figures such as:
# 1) petal_lenght vs. petal_width
# 2) petal_length and sepal_width
# find the means of each specimens in each graph and calculate the distance between the means and draw a line
# with a text showing the distance use plt.text

plt.rcParams["figure.figsize"] = [12,8]

plt.subplot(2,2,1)
plt.xlabel('Petal Length')
plt.ylabel('Sepal Width')
plt.title('Petal Length vs Sepal Width')
plt.scatter(setosa_petal_len, setosa_sepal_wid, c="b", label ='setosa')
plt.scatter(versi_petal_len, versi_sepal_wid, c="darkorange", label ='versicolor')

#plots the mean of each species
plt.plot(setosa_petal_len_mean,setosa_sepal_wid_mean, c='b', marker='+')
plt.plot(versi_petal_len_mean,versi_sepal_wid_mean, c='darkorange', marker='+')

#plots the regression line
plt.plot([setosa_petal_len_mean,versi_petal_len_mean],[setosa_sepal_wid_mean,versi_sepal_wid_mean],'r-')
plt.text(3,3.1,dist_versi_setosa1)
plt.legend(loc='upper right')


plt.subplot(2,2,2)
plt.xlabel('Petal Length')
plt.ylabel('Petal Width')
plt.title('Petal Length vs Petal Width')
plt.scatter(setosa_petal_len, setosa_petal_wid, c="b", label ='setosa')
plt.scatter(versi_petal_len, versi_petal_wid, c="darkorange", label ='versicolor')

#plots the mean of each species
plt.plot(setosa_petal_len_mean,setosa_petal_wid_mean, c='b', marker='+')
plt.plot(versi_petal_len_mean,versi_petal_wid_mean, c='darkorange', marker='+')

#plots the regression line
plt.plot([setosa_petal_len_mean,versi_petal_len_mean],[setosa_petal_wid_mean,versi_petal_wid_mean],'r-')
plt.text(3,0.75,dist_versi_setosa2)
plt.legend(loc='upper right')


# species = iris_data_2.groupby("species")

# for name, specie in species:
#     plt.xlabel('Petal Length')
#     plt.ylabel('Sepal Width')
#     plt.title('Petal Length vs Sepal Width')
#     plt.subplot(2,2,1)
#     plt.plot(specie["petal_length"], specie["sepal_width"],'.', label=name)
#     plt.legend() 
    
#     plt.xlabel('Petal Length')
#     plt.ylabel('Petal Width')
#     plt.title('Petal Length vs Petal Width')
#     plt.subplot(2,2,2)
#     plt.plot(specie["petal_length"], specie["petal_width"],'.',label=name)
#     plt.legend()

# print(species)
```

```python
# use seaborn's jointplot to show the regression line
sns.jointplot(x='petal_length', y='sepal_width', data=iris_data_2, kind = 'reg')

sns.jointplot(x='petal_length', y='petal_width', data=iris_data_2, kind = 'reg')
```

```python
# use the given regression code in the class to find the regression line for only setosa for petal_length and petal_width

x = setosa_petal_len
y = setosa_petal_wid
z = np.polyfit(x, y, 1)

# create polynomial
p = np.poly1d(z)
print(p)

# Extrapolation
x0=2
y0=np.polyval(p,x0)
plt.scatter(x0,y0)
print(y0)

# plot polynomial
xp = np.linspace(1, 2, 100)
plt.plot(x, y, '.', xp, p(xp), '-')

plt.xlabel('Petal Length')
plt.ylabel('Petal Width')
plt.title('Regression line for Petal Length vs Petal Width for Setosa specie')

plt.show()
```
